```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.linspace(0, 2 * np.pi, 1000)

first_equation = 5.5*np.cos(2*arr)+5.5
second_equation = 0.02 * np.exp(arr)
third_equation = 0.25 * arr**2 + 0.1 * np.sin(10*arr)

plt.ylim(-1, 10)
plt.plot(arr, first_equation)
plt.plot(arr, second_equation)
plt.plot(arr, third_equation)
```
